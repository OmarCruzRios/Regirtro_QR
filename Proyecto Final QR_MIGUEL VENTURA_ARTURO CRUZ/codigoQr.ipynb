```python
import tkinter as tk
#importar los modulos restantes de tkinter 
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
#from pyzbar.pyzbar import decode
# modulos creados 
from usuarios import *
from Conexion1 import *
```

```python

class FormularioClientes:
    
    # valores para ser modificados
        
    
        global base 
        base =None
        
        global groupBox
        groupBox =None
        
        global texBoxNumeroRegistro
        texBoxNumeroRegistro =None
    
        global texBoxNumeroCuenta 
        texBoxNumeroCuenta =None
    
        global texBoxNombreCompleto  
        texBoxNombreCompleto =None  
    
        global texBoxCelular  
        texboxCelular =None   
    
        global combo
        combo =None
    
        global tree
        tree =None
    
def  Formulario():
    global base
    global groupBox
    global texBoxNumeroRegistro
    global texBoxNumeroCuenta
    global texBoxNombreCompleto
    global texBoxCelular
    global combo
    global tree
    
    try:
            base = Tk()
            base.geometry("1600x350")
            base.title("REGISTRO DE ALUMNOS")
            
            groupBox =LabelFrame(base, text="Datos de usuarios ",padx=10, pady=10)
            groupBox.grid(row=0,column=0,padx=10,pady=10)
            
            labelNumeroRegistro=Label(groupBox,text="N° de registro: ", width=13, font=("arial",12)).grid(row=0,column=0)
            texBoxNumeroRegistro = Entry(groupBox)
            texBoxNumeroRegistro.grid(row=0, column=1)
            
            labelNumeroCuenta=Label(groupBox,text="N° de cuenta: ", width=13, font=("arial",12)).grid(row=1,column=0)
            texBoxNumeroCuenta = Entry(groupBox)
            texBoxNumeroCuenta.grid(row=1, column=1)
            
            labelNombreCompleto=Label(groupBox,text="Nombre completo: ", width=13, font=("arial",12)).grid(row=2,column=0)
            texBoxNombreCompleto = Entry(groupBox)
            texBoxNombreCompleto.grid(row=2, column=1)
            
            labelCelular=Label(groupBox,text="Celular: ", width=13, font=("arial",12)).grid(row=3,column=0)
            texBoxCelular= Entry(groupBox)
            texBoxCelular.grid(row=3, column=1)
            
            labelSexo=Label(groupBox,text="Sexo: ", width=13, font=("arial",12)).grid(row=4,column=0)
            seleccionSexo = tk.StringVar()
            combo=ttk.Combobox(groupBox,values=["Masculino","Femenino"],textvariable=seleccionSexo)
            combo.grid(row=4,column=1)
            #seleccionSexo.set("Femenino")
            
            #agregar la opcion de correo electronico pendiente
            
            Button(groupBox,text="Guardar",width=10,command=guardarRegistros).grid(row=6, column=0)
            Button(groupBox,text="Modificar",width=10,command=modificarRegistros).grid(row=6, column=1)
            Button(groupBox,text="Eliminar",width=10,command=eliminarRegistros).grid(row=6, column=2)
            Button(groupBox,text="generarQR",width=10,command=qrRegistros).grid(row=7, column=0)
           # Button(groupBox,text="mostrarlectorQR",width=10,command=capturarRegistros).grid(row=7, column=1)
            
            
            # Lista de personas registradas
            
            groupBox = LabelFrame(base,text="Lista de personas registradas", padx=5,pady=5,)
            groupBox.grid(row=0,column=1,padx=5,pady=5)
            
            #crear un treeview
            #configurar las columnas 
            
            tree =ttk.Treeview(groupBox,columns=("N_de_registro","N_de_cuenta", "nombre completo", "celular","sexo"),show='headings',height=5,)
            tree.column("#1",anchor=CENTER)
            tree.heading("#1",text="N_de_registro")
            
            tree.column("#2",anchor=CENTER)
            tree.heading("#2",text="N_de_cuenta")
            
            tree.column("#3",anchor=CENTER)
            tree.heading("#3",text="nombre completo")
            
            tree.column("#4",anchor=CENTER)
            tree.heading("#4",text="celular")
            
            tree.column("#5",anchor=CENTER)
            tree.heading("#5",text="sexo")
            
           
            # agregar los datos a la tabla 
            # mostrar la tabla
            
            for row in CClientes.mostrarClientes():
                tree.insert("","end",values=row)
                
            # ejecutar la funcion de hacer click y mostrar el resultado en los entry
            # de modificar
            
            tree.bind("<<TreeviewSelect>>",seleccionarRegistro)
        
            tree.pack()
            base.mainloop()
    except ValueError as error:
            print("Error al mostrar la interfaz,error: {}".format(error))
                
def guardarRegistros():
    
    global texBoxNumeroCuenta
    global texBoxNombreCompleto
    global texBoxCelular
    global combo
    global groupBox
        
    try:
            #veryficar si los widgets estan inicializados
            if texBoxNumeroCuenta is None or texBoxNombreCompleto is None or texBoxCelular is None or combo is None:
                print("Los widgets no estan inicializados")
                return
            
            N_de_cuenta = texBoxNumeroCuenta.get()
            nombre = texBoxNombreCompleto.get()
            celular = texBoxCelular.get()
            sexo = combo.get()
            
            CClientes.ingresarClientes(N_de_cuenta, nombre, celular, sexo) #solo se modifico celular
            messagebox.showinfo("informacion","los datos fueron guardados")
            
            actualizarTreeView()
            
            #limpiamos los campos 
            
            texBoxNumeroCuenta.delete(0,END)
            texBoxNombreCompleto.delete(0,END)
            texBoxCelular.delete(0,END)
            
    except ValueError as error:
            print("Error al ingresar los datos {}".format(error))
            

def actualizarTreeView():
    global tree 
    
    try:
        #borra todos los elementos actuales del treeview
        tree.delete(*tree.get_children())
        # obtener los nuevos datos que deseamos mostrar
        datos = CClientes.mostrarClientes
        # insertar los nuevos datos en el treeview
        for row in CClientes.mostrarClientes():
                tree.insert("","end",values=row)
    except ValueError as error:
        print("Error al actualizar tabla{}".format(error))
        
def seleccionarRegistro(event):
    try:
        itemSeleccionado = tree.focus()
        
        if itemSeleccionado:
            # obtener los valores por columna
            values = tree.item(itemSeleccionado)['values']
            # establecer los valores en los widgets Entry
            texBoxNumeroRegistro.delete(0,END)       
            texBoxNumeroRegistro.insert(0,values[0])  
            
            texBoxNumeroCuenta.delete(0,END)          
            texBoxNumeroCuenta.insert(0,values[1])    
            
            texBoxNombreCompleto.delete(0,END)        
            texBoxNombreCompleto.insert(0,values[2])   
            
            texBoxCelular.delete(0,END)
            texBoxCelular.insert(0,values[3])
            
            combo.set(values[4])
            
    except ValueError as error:
        print("Error al seleccionar registro {}".format(error))
        
def modificarRegistros():
    
    global texBoxNumeroRegistro
    global texBoxNumeroCuenta
    global texBoxNombreCompleto
    global texBoxCelular
    global combo
    global groupBox
        
    try:
            #veryficar si los widgets estan inicializados
            if texBoxNumeroRegistro is None or texBoxNumeroCuenta is None or texBoxNombreCompleto is None or texBoxCelular is None or combo is None:
                print("Los widgets no estan inicializados")
                return
            N_de_registro= texBoxNumeroRegistro.get()   
            N_de_cuenta = texBoxNumeroCuenta.get()   
            nombre = texBoxNombreCompleto.get()   
            celular = texBoxCelular.get()    
            sexo = combo.get()
            
            CClientes.modificarClientes(N_de_registro,N_de_cuenta, nombre, celular, sexo)
            messagebox.showinfo("informacion","los datos fueron actualizados")
            
            actualizarTreeView()
            
            #limpiamos los campos 
            texBoxNumeroRegistro.delete(0,END)
            texBoxNumeroCuenta.delete(0,END)
            texBoxNombreCompleto.delete(0,END)
            texBoxCelular.delete(0,END)
            
    except ValueError as error:
            print("Error al modificar los datos {}".format(error))
            
def eliminarRegistros():
    
    global texBoxNumeroRegistro
    global texBoxNumeroCuenta
    global texBoxNombreCompleto
    global texBoxCelular
        
    try:
            #veryficar si los widgets estan inicializados
            if texBoxNumeroRegistro is None: # texBoxNumeroCuenta
                print("Los widgets no estan inicializados")
                return
            
            N_de_registro = texBoxNumeroRegistro.get()
    
            CClientes.eliminarClientes(N_de_registro)
            messagebox.showinfo("informacion","los datos fueron eliminados")
            
            actualizarTreeView()
            
            #limpiamos los campos 
            texBoxNumeroRegistro.delete(0,END)
            texBoxNumeroCuenta.delete(0,END)
            texBoxNombreCompleto.delete(0,END)
            texBoxCelular.delete(0,END)
            
    except ValueError as error:
            print("Error al eliminar los datos {}".format(error))
            

        
        
        #obtener el numero de cuenta 
        #prueba de generar y guardar qr
        # arreglar errores e inconsistencias
def qrRegistros():
    
    global texBoxNumeroRegistro
    global texBoxNumeroCuenta
    global texBoxNombreCompleto
    global texBoxCelular
    global combo
    global groupBox
        
    try:
            #veryficar si los widgets estan inicializados
            if texBoxNumeroRegistro is None or texBoxNumeroCuenta is None or texBoxNombreCompleto is None or texBoxCelular is None or combo is None:
                print("Los widgets no estan inicializados")
                return
            N_de_registro= texBoxNumeroRegistro.get()
            N_de_cuenta = texBoxNumeroCuenta.get()
            nombre = texBoxNombreCompleto.get()
            celular = texBoxCelular.get()
            sexo = combo.get()
            
            CClientes.crearqr(N_de_registro,N_de_cuenta, nombre, celular, sexo)
            messagebox.showinfo("informacion","codigo QR creado")
            
            actualizarTreeView()
            
            #limpiamos los campos 
            texBoxNumeroRegistro.delete(0,END)
            texBoxNumeroCuenta.delete(0,END)
            texBoxNombreCompleto.delete(0,END)
            texBoxCelular.delete(0,END)
            
    except ValueError as error:
            print("Error al crear QR{}".format(error))

            
            
    #lector de qr
    

            

Formulario()


# me falta el autocompletar
```
